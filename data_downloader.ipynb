```python
import requests
from bs4 import BeautifulSoup
import os
import csv

os.makedirs('downloaded_images', exist_ok=True)

with open('image_data.csv', 'w', newline='') as csvfile:
    fieldnames = ['image_name', 'date']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()

    num_pages = 2000

    for page in range(1, num_pages + 1):
        url = f'https://www.musinsa.com/mz/brandsnap?_m=&gender=&mod=&bid=&p={page}'
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        list_items = soup.find_all(class_='listItem')

        for index, item in enumerate(list_items):
            img_tag = item.find('img')
            date_div = item.find(class_='division')

            if img_tag and date_div:
                img_url = img_tag['src']
                date = date_div.text.strip()
            else:
                continue

            img_response = requests.get(img_url)
            
            img_file_name = f'image_{page}_{index}.jpg'
            
            with open(f'downloaded_images/{img_file_name}', 'wb') as file:
                file.write(img_response.content)
            
            writer.writerow({'image_name': img_file_name, 'date': date})

            print(f"Saved data (Page {page}, Item {index}): ", img_file_name, date)
```
